```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
Empll=pd.read_csv('/content/zill.csv')
X=Empll[['IncidentsTotal']]
y=Empll[['SeverityScore']]
regressor = LinearRegression()
regressor.fit(X, y)
Empll.plot(x='IncidentsTotal', y='IncidentsTotal', style='o', label='Severity Point')
plt.plot(X,regressor.predict(X),color='red',linewidth=2,label='Regression Line')
plt.xlabel('IncidentsTotal')
plt.ylabel('SeverityScore')
plt.legend()
plt.show()
```
